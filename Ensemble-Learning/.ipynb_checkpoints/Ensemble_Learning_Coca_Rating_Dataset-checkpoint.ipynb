## Univariate analysis

```python
plt.figure(figsize=(25,20))
plt.subplot(2,2,1)
sns.histplot(df['Review'], kde=True)

plt.subplot(2,2,2)
sns.histplot(df['Cocoa_Percent'], kde=True)

plt.subplot(2,2,3)
sns.histplot(df['Rating'], kde=True)

plt.show()
```

```python
df.skew().sort_values(ascending=False)
# from the above histogram Cocoa_Percent is right skewed with skewness 1.05 and Review and Rating are 
# left skewed with -0.527929, -0.581678,respectivly.
```

```python
# so use log transformation to make them normalized
df["Cocoa_Percent"]=np.log(df["Cocoa_Percent"])
```

```python
sns.histplot(df['Cocoa_Percent'], kde=True)
plt.show()
print("Skewness after transformation:",df["Cocoa_Percent"].skew())
```

```python
plt.figure(figsize=(25,20))
plt.subplot(2,2,1)
sns.boxplot(df['Review'])
plt.xlabel("Review")

plt.subplot(2,2,2)
sns.boxplot(df['Cocoa_Percent'])
plt.xlabel("Cocoa_Percent")

plt.subplot(2,2,3)
sns.boxplot(df['Rating'])
plt.xlabel("Rating")

plt.show()
```

```python
# from above boxplot, we can see that Rating and Cocoa_Percent columns have outliers
# With help of winsorize, remove the outliers
winsor=Winsorizer(capping_method="iqr",fold=1.5,tail="both",variables=["Cocoa_Percent"])
df["Cocoa_Percent"]=winsor.fit_transform(df[["Cocoa_Percent"]])
```

```python
winsor=Winsorizer(capping_method="iqr",fold=1.5,tail="both",variables=["Rating"])
df["Rating"]=winsor.fit_transform(df[["Rating"]])
```

```python
# After removal of outliers, the boxplot are:
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
sns.boxplot(df['Cocoa_Percent'])
plt.xlabel("Cocoa_Percent")

plt.subplot(1,2,2)
sns.boxplot(df['Rating'])
plt.xlabel("Rating")

plt.show()
```

## Bivariate analysis

```python
sns.heatmap(df.corr(),cmap="coolwarm",annot=True,fmt=".02f")
plt.show()
```

```python
# Now do the label encoding to the object data
le=LabelEncoder()
df["Company"]=le.fit_transform(df["Company"])
df["Company_Location"]=le.fit_transform(df["Company_Location"])
df["Bean_Type"]=le.fit_transform(df["Bean_Type"])
df["Origin"]=le.fit_transform(df["Origin"])
```

```python
df.head()
```

```python
x=df.drop(["Rating"],axis=1)
y=df["Rating"]
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
x_scaled
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,stratify=y,random_state=10)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

## Train a model using decision tree and then using bagging

```python
from sklearn.metrics import mean_squared_error
base_estimator = DecisionTreeRegressor()

# Initialize the BaggingRegressor
bagging_regressor = BaggingRegressor(base_estimator=base_estimator, n_estimators=10, random_state=42)

# Train the BaggingRegressor
bagging_regressor.fit(x_train, y_train)

# Predict ratings for the test set
y_pred = bagging_regressor.predict(x_test)

# Evaluate the performance (e.g., using Mean Squared Error)
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```
