```python
# Removal missing values by imputation
mean_Schooling=np.mean(df["Schooling"])
mean_Income_composition=np.mean(df["Income_composition"])
mean_Population=np.mean(df["Population"])
mean_GDP=np.mean(df["GDP"])
mean_Total_expenditure=np.mean(df["Total_expenditure"])
mean_Hepatitis_B=np.mean(df["Hepatitis_B"])
mean_Alcohol=np.mean(df["Alcohol"])
```

```python
df["Schooling"]=df["Schooling"].fillna(mean_Schooling)
df["Income_composition"]=df["Income_composition"].fillna(mean_Income_composition)
df["Population"]=df["Population"].fillna(mean_Population)
df["GDP"]=df["GDP"].fillna(mean_GDP)
df["Total_expenditure"]=df["Total_expenditure"].fillna(mean_Total_expenditure)
df["Hepatitis_B"]=df["Hepatitis_B"].fillna(mean_Hepatitis_B)
df["Schooling"]=df["Schooling"].fillna(mean_Schooling)
df["Alcohol"]=df["Alcohol"].fillna(mean_Alcohol)
```

```python
df.dropna(inplace=True)
```

```python
x=df.drop(["Life_expectancy"],axis=1)
y=df["Life_expectancy"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
r2=r2_score(y_test,y_pred)
r2
```

# Apply ridge and lasso regression

```python
ridge_alpha=1.0
lasso_alpha=0.01
```

```python
ridge=Ridge(alpha=ridge_alpha)
lasso=Lasso(alpha=lasso_alpha)
```

```python
ridge.fit(x_train,y_train)
lasso.fit(x_train,y_train)
```

```python
ridge_pred=model.predict(x_test)
lasso_pred=model.predict(x_test)
```

```python
r2_lasso=r2_score(y_test,lasso_pred)
print(r2_lasso)
```

```python
r2_ridge=r2_score(y_test,ridge_pred)
print(r2_ridge)
```
